```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# 1. Generate Simple Simulated Dataset
np.random.seed(42)
n = 300
df = pd.DataFrame({
    'Department': np.random.choice(['Sales', 'Eng', 'HR', 'Mkt', 'Fin'], n),
    'Salary': np.random.choice(['Low', 'Medium', 'High'], n),
    'Experience': np.random.randint(1, 15, n),
    'Satisfaction': np.random.uniform(0.1, 1.0, n),
    'Attrition': np.random.choice([1, 0], n, p=[0.2, 0.8])
})

# 2. Configure Dashboard Layout (Compact Size: 10x8)
fig, axes = plt.subplots(2, 2, figsize=(6, 5))
fig.suptitle('HR Employee Attrition Dashboard', fontsize=16, fontweight='bold')

# Chart 1: Overall Attrition Rate (Pie Chart)
df['Attrition'].value_counts().plot.pie(ax=axes[0, 0], autopct='%1.1f%%', colors=['#4C72B0', '#C44E52'], labels=['No', 'Yes'])
axes[0, 0].set_title('Overall Attrition Rate')
axes[0, 0].set_ylabel('')

# Chart 2: Attrition Rate by Department (Bar Graph)
sns.barplot(x='Department', y='Attrition', data=df, ax=axes[0, 1], ci=None, palette='viridis')
axes[0, 1].set_title('Attrition by Department')
axes[0, 1].set_ylabel('Attrition Rate')

# Chart 3: Attrition Rate by Salary Tier (Bar Graph)
sns.barplot(x='Salary', y='Attrition', data=df, ax=axes[1, 0], ci=None, order=['Low', 'Medium', 'High'], palette='coolwarm')
axes[1, 0].set_title('Attrition by Salary Tier')
axes[1, 0].set_ylabel('Attrition Rate')

# Chart 4: Attrition by Experience & Satisfaction (Scatter Plot)
sns.scatterplot(x='Experience', y='Satisfaction', hue='Attrition', data=df, ax=axes[1, 1], palette={1: '#C44E52', 0: '#4C72B0'}, alpha=0.7)
axes[1, 1].set_title('Attrition by Exp & Satisfaction')

plt.tight_layout()
plt.show()
```
